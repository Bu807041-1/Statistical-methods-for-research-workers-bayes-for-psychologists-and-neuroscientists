```python
One_samp_t_estimation = """

data{

int<lower=0> N;
vector[N] Y;

}

parameters{

 real<lower=0, upper=1> mu;
 real<lower=0> sigma; 
 
}

model{
//priors
mu ~ normal(0.5, 0.2);

// Likliehood
Y ~ normal(mu, sigma);

}

generated quantities {
  real ES = (mu - 0.5) / sigma; //effect size
}

"""
```

```python
Baseline_prop = red_df["Baseline_Proportion_Gaze_to_Singer"].to_numpy()

One_sample_dat = {'N': len(Baseline_prop),
                  'Y': Baseline_prop}

fit_One = ps.stan(model_code = One_samp_t_estimation, data = One_sample_dat, iter=10000, chains=4)
```

```python
print(fit_One)
samples = fit_One.extract(permuted=True)
mu_mean = stats.mean(samples['mu'])
mu_mean
```
