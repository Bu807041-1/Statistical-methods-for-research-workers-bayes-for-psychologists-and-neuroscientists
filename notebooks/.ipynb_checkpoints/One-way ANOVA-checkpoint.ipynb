# Treament coding 
 Generating a design (contrast) matrix for categorical regression 
 analysis with treatment coding as aOne way ANOVA analyis

```python
df.Condition = df.Condition.astype("category")
df.Condition = df['Condition'].cat.reorder_categories([2, 1,3,4])
Reordered =  dmatrix("Condition", df)
Reordered = np.array(Reordered)

#Droping the intercept term
Reordered = Reordered[:,1:]

# check design matrix is coded approapriately.
Reordered
```

```python
One_wayANOVA = """
data{
int<lower=0> Rows; // Number of rows in the contrast matix
int<lower=0> Columns; // Number columns

vector[Rows] DV; // Dependent variable data points

matrix[Rows,Columns] x; //Contrast matrix
}

parameters{
real<lower=0> alpha; // intercept

vector[Columns] coef; // Beta coefficients 

real<lower=0> sigma;

}
model{
//No priorss set so results will be similar to maximum liklihood estimates 


//likelihood
DV ~ normal(alpha + x*coef  , sigma);
}
"""
```

```python
sm = ps.StanModel(model_code = One_wayANOVA)
```

```python
#Generate python dictionary for passing the relevant data for the analysis to stan

data = {
    
    'Rows': len(df),
    #
      'Columns': Reordered.shape[1],
    
       'DV': df["Days_One_to_Seven_Number_of_Intrusions"],
    
        'x':Reordered}
```

```python
fit = sm.sampling(data = data, iter = 10000, chains=4, seed= 302675)
```

```python
az.plot_trace(fit)
```

```python
print(fit);
```
