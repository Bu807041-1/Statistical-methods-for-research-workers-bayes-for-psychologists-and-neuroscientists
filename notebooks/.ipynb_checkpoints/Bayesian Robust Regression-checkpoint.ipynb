Teh data example below is from https://stats.idre.ucla.edu/r/dae/robust-regression/. Teh code below is for conducting Robust regression which differs based on the use of student-t distrtibution as the likelihood function to model the data and accomafate outliers.

```python
url = "https://stats.idre.ucla.edu/stat/data/crime.dta"
df = pd.read_stata(url)
df.head()
```

```python
sns.distplot(df["crime"])
plt.show()

sns.distplot(df["poverty"])
plt.show()


sns.distplot(df["single"])
plt.show
```

```python
Robust_Regression = """
data {
  int<lower=0> N;   // Sample size
  int<lower=0> K;   // Number of predictor variables
  matrix[N, K] x;   // Predictor variable (IV) matrix
  vector[N] y;      // Vector of Dependent variable (DV) values
}
parameters {
  real nu
  real alpha;           // Intercept term
  vector[K] beta;       // Coefficients for predictors
  real<lower=0> sigma;  // Standard deviation
}
model {

//priors

  y ~ student_t(nu, alpha + x * beta, sigma);  // likelihood
}
"""
```
